Question

Which game genres appear most frequently?

```python
import plotly.io as pio

pio.renderers.default = "notebook_connected"

print(pio.renderers.default)
```

```python
genre_count = (
    df.groupby("Genre")
      .size()
      .reset_index(name="Games")
      .sort_values("Games", ascending=False)
)

fig = px.bar(
    genre_count,
    x="Genre",
    y="Games",
    color="Games",
    title="Top Game Genres by Number of Games",
    text_auto=True
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Genre",
    yaxis_title="Number of Games"
)

fig.show()
```

```python
## =====================================================
# EDA 1 - Distribution of Video Game Genres
# =====================================================

# Create a clean copy of the dataset (run this once)
df_clean = df.copy()

# Count the number of games in each genre
genre_count = (
    df_clean["Genre"]
    .value_counts()
    .reset_index()
)

# Rename the columns
genre_count.columns = ["Genre", "Number of Games"]

# Create the bar chart
fig = px.bar(
    genre_count,
    x="Genre",
    y="Number of Games",
    color="Number of Games",
    color_continuous_scale="Blues",
    text="Number of Games",
    title="Distribution of Video Game Genres"
)

# Display values on top of the bars
fig.update_traces(textposition="outside")

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Distribution of Video Game Genres",
        "x": 0.5,
        "xanchor": "center"
    },
    xaxis_title="Genre",
    yaxis_title="Number of Games",
    coloraxis_showscale=False,
    height=600,
    width=1000
)

# Show the chart
fig.show()
```

# Question 3

## Business Question
Which video game genres generate the highest global sales?

### Why is this important?
This analysis helps publishers and game developers understand which genres are the most profitable. It provides insights into consumer preferences and supports decisions related to future game development and investment.

```python
# =====================================================
# QUESTION 3
# Which Genres Generate the Highest Global Sales?
# =====================================================

# Calculate total global sales by genre
genre_sales = (
    df_clean.groupby("Genre")["Global_Sales"]
    .sum()
    .reset_index()
    .sort_values(by="Global_Sales", ascending=False)
)

# Create bar chart
fig = px.bar(
    genre_sales,
    x="Genre",
    y="Global_Sales",
    color="Global_Sales",
    color_continuous_scale="Plasma",
    text="Global_Sales",
    title="Total Global Sales by Video Game Genre"
)

# Format text labels
fig.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Total Global Sales by Video Game Genre",
        "x": 0.5,
        "xanchor": "center"
    },
    xaxis_title="Genre",
    yaxis_title="Global Sales (Millions)",
    coloraxis_showscale=False,
    height=600,
    width=1000
)

fig.show()
```

### Insight

Action games generate the highest global sales, followed by Sports and Shooter games. These genres have consistently attracted a large player base across different gaming platforms and regions. On the other hand, Puzzle and Strategy games contribute comparatively lower global sales.

## Business Question

# Question 5

## Business Question
How has the number of video game releases changed over the years?

### Why is this important?

This analysis helps identify trends in the gaming industry by showing periods of growth and decline in game releases. It provides insights into market expansion, technological advancements, and changes in consumer demand over time.

```python
# =====================================================
# QUESTION 5
# How Has the Number of Video Game Releases Changed Over the Years?
# =====================================================

# Count the number of games released each year
games_per_year = (
    df_clean.groupby("Year_of_Release")
    .size()
    .reset_index(name="Number of Games")
)

# Create line chart
fig = px.line(
    games_per_year,
    x="Year_of_Release",
    y="Number of Games",
    markers=True,
    title="Number of Video Game Releases by Year"
)

# Improve line appearance
fig.update_traces(
    line=dict(width=3),
    marker=dict(size=7)
)

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Number of Video Game Releases by Year",
        "x": 0.5,
        "xanchor": "center"
    },
    xaxis_title="Year",
    yaxis_title="Number of Games Released",
    height=600,
    width=1000
)

fig.show()
```

### Insight

The number of video game releases increased steadily from the early 1980s and experienced rapid growth during the late 1990s and 2000s. The highest number of game releases occurred around the late 2000s, after which the number of releases gradually declined. This trend reflects the expansion of the gaming industry followed by market shifts toward digital distribution and newer gaming technologies.

# Question 7

## Business Question
How do video game genre preferences differ across North America, Europe, Japan, and Other Regions?

### Why is this important?

Consumer preferences vary across different regions. Understanding these differences helps game publishers and developers create region-specific marketing strategies, localize content, and make better business decisions.

```python
# =====================================================
# QUESTION 7
# Genre Preferences Across Different Regions
# =====================================================

# Calculate total sales by genre for each region
genre_region_sales = (
    df_clean.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]
    .sum()
    .reset_index()
)

# Convert from wide format to long format
genre_region_sales = genre_region_sales.melt(
    id_vars="Genre",
    var_name="Region",
    value_name="Sales"
)

# Rename region names for better readability
genre_region_sales["Region"] = genre_region_sales["Region"].replace({
    "NA_Sales": "North America",
    "EU_Sales": "Europe",
    "JP_Sales": "Japan",
    "Other_Sales": "Other Regions"
})

# Create grouped bar chart
fig = px.bar(
    genre_region_sales,
    x="Genre",
    y="Sales",
    color="Region",
    barmode="group",
    text="Sales",
    title="Video Game Genre Preferences Across Different Regions"
)

# Format labels
fig.update_traces(
    texttemplate="%{text:.1f}",
    textposition="outside"
)

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Video Game Genre Preferences Across Different Regions",
        "x": 0.5,
        "xanchor": "center"
    },
    xaxis_title="Genre",
    yaxis_title="Sales (Millions)",
    legend_title="Region",
    height=700,
    width=1200
)

fig.show()
```

### Insight

The analysis shows that Action and Sports games are highly popular in North America and Europe, generating the highest sales in these regions. In contrast, Japan shows a stronger preference for Role-Playing games compared to other regions. These regional differences highlight the importance of adapting game development and marketing strategies to local consumer preferences.

# Question 8

## Business Question
Is there a relationship between critic scores and global sales of video games?

### Why is this important?

Critic reviews can influence purchasing decisions and affect a game's commercial success. This analysis helps determine whether higher critic ratings are associated with higher global sales.

```python
# =====================================================
# QUESTION 8
# Relationship Between Critic Score and Global Sales
# =====================================================

# Create dataset with required columns only
critic_sales = df_clean[["Critic_Score", "Global_Sales", "Genre"]].dropna()

# Create scatter plot
fig = px.scatter(
    critic_sales,
    x="Critic_Score",
    y="Global_Sales",
    color="Genre",
    opacity=0.7,
    title="Relationship Between Critic Score and Global Sales",
    labels={
        "Critic_Score": "Critic Score",
        "Global_Sales": "Global Sales (Millions)"
    },
    hover_data=["Genre"]
)

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Relationship Between Critic Score and Global Sales",
        "x": 0.5,
        "xanchor": "center"
    },
    height=650,
    width=1100
)

fig.show()
```

```python
# =====================================================
# Correlation Between Critic Score and Global Sales
# =====================================================

correlation = critic_sales["Critic_Score"].corr(critic_sales["Global_Sales"])

print(f"Correlation between Critic Score and Global Sales: {correlation:.2f}")
```

### Insight

The scatter plot shows a positive relationship between critic scores and global sales. Games with higher critic ratings generally tend to achieve better sales, although there are several exceptions. The calculated correlation coefficient provides a quantitative measure of this relationship, indicating that critic reviews can influence commercial success but are not the only factor affecting sales.

# Question 9

## Business Question
Is there a relationship between user scores and global sales of video games?

### Why is this important?

User scores reflect player satisfaction and overall gaming experience. This analysis helps determine whether games that receive higher ratings from players tend to achieve greater commercial success.

```python
# =====================================================
# QUESTION 9
# Relationship Between User Score and Global Sales
# =====================================================

# Select required columns and remove missing values
user_sales = df_clean[["User_Score", "Global_Sales", "Genre"]].dropna()

# Convert User_Score to numeric
user_sales["User_Score"] = pd.to_numeric(
    user_sales["User_Score"],
    errors="coerce"
)

# Remove rows that became NaN after conversion
user_sales = user_sales.dropna()

# Create scatter plot
fig = px.scatter(
    user_sales,
    x="User_Score",
    y="Global_Sales",
    color="Genre",
    opacity=0.7,
    title="Relationship Between User Score and Global Sales",
    labels={
        "User_Score": "User Score",
        "Global_Sales": "Global Sales (Millions)"
    },
    hover_data=["Genre"]
)

fig.update_layout(
    template="plotly_white",
    title={
        "text": "Relationship Between User Score and Global Sales",
        "x": 0.5,
        "xanchor": "center"
    },
    height=650,
    width=1100
)

fig.show()
```

```python
# =====================================================
# Correlation Between User Score and Global Sales
# =====================================================

correlation = user_sales["User_Score"].corr(user_sales["Global_Sales"])

print(f"Correlation between User Score and Global Sales: {correlation:.2f}")
```

### Insight

The scatter plot indicates a weak to moderate positive relationship between user scores and global sales. While many highly rated games achieve strong sales, there are also games with excellent user ratings but relatively low sales. This suggests that player satisfaction contributes to commercial success, but factors such as marketing, platform popularity, release timing, and brand recognition also influence overall sales.

# Question 10

## Business Question
Which ESRB (Age) ratings generate the highest global sales?

### Why is this important?

Age ratings classify games according to their target audience. Analyzing global sales by ESRB rating helps publishers understand which audience segments contribute the most to commercial success and supports future game development and marketing strategies.

```python
# =====================================================
# QUESTION 10
# Which ESRB Ratings Generate the Highest Global Sales?
# =====================================================

# Calculate total global sales by ESRB rating
rating_sales = (
    df_clean.groupby("Rating")["Global_Sales"]
    .sum()
    .reset_index()
    .sort_values(by="Global_Sales", ascending=False)
)

# Create bar chart
fig = px.bar(
    rating_sales,
    x="Rating",
    y="Global_Sales",
    color="Global_Sales",
    color_continuous_scale="Cividis",
    text="Global_Sales",
    title="Global Sales by ESRB Rating"
)

# Format text labels
fig.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

# Improve layout
fig.update_layout(
    template="plotly_white",
    title={
        "text": "Global Sales by ESRB Rating",
        "x": 0.5,
        "xanchor": "center"
    },
    xaxis_title="ESRB Rating",
    yaxis_title="Global Sales (Millions)",
    coloraxis_showscale=False,
    height=600,
    width=1000
)

fig.show()
```

### Insight

The analysis shows that games rated **E (Everyone)** and **T (Teen)** generate the highest global sales, indicating strong demand from broad and younger audiences. **M (Mature)** rated games also contribute significantly to sales, while ratings such as **AO (Adults Only)** and **EC (Early Childhood)** account for a much smaller share of the market. These findings suggest that games targeting wider audiences generally achieve greater commercial success.
